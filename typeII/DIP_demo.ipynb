```python
def normalization_x(filename):
    # 徐龙老师采用的消除横线噪声的通道归一化方法
    image = filename.copy()
    # list 没有shape方法
    shape = image.shape
    global_mean = np.mean(image)
    # 防止255溢出，先将数组转换成列表
    image = image.tolist()
    for i in range(shape[0]):
        channel_mean = np.mean(image[i])
        image[i] = image[i] - channel_mean + global_mean
    image = np.array(image, dtype=np.uint8)
    return image


def normalization_z(filename):
    # 张沛锦采用的消除背景缓变噪声的通道归一化方法
    image = filename.copy()
    # list 没有shape方法
    shape = image.shape
    # 防止255溢出，先将数组转换成列表
    image = image.tolist()

    for i in range(shape[0]):
        channel_mean = np.mean(image[i])
        image[i] = abs(image[i] - channel_mean)
    image = np.array(image, dtype=np.uint8)
    return image
```

```python
# 两种通道归一化
image_x = normalization_x(image)
image_z = normalization_z(image)
images = np.hstack([image, image_x, image_z])

cv.imshow("normalization", images)
cv.waitKey()
cv.destroyWindow("normalization")
```

```python
#自适应二值化

image_z_b = cv.adaptiveThreshold(image_z,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,9,0)
image_x_b = cv.adaptiveThreshold(image_x,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,9,0)



images = np.hstack([image_z_b, image_x_b])
cv.imshow("binary", images)
cv.waitKey()
cv.destroyWindow("binary")
```
